## error bar

10k : (1) 6min 12s (2) 6min 24s (3) 6min 2s  
30k : (1) 13min 17s (2) 12min 53s  (3) 13min 45s  
50k :  (1) 24min 43s (2) 26min 16s (3) 27min 13s  
100k : (1) 37min 27s (2) 38min 5s (3) 39min 14s  
500k : (1) 9h 27min 19s (2) 8h 15min 6s (3) 7h 14min 30s  
1m : (1) 13h 46m 33s (2) 13h 46m 42s (3) 14h 17m 38s  

```python
import numpy as np
import matplotlib.pyplot as plt

times = np.array([
    [0.1033, 0.1067, 0.1006],
    [0.2214, 0.2153, 0.2292],
    [0.4120, 0.4378, 0.4536],
    [0.6242, 0.6347, 0.6544],
    [9.4553, 8.2517, 7.2417],
    [13.7758, 13.7783, 14.2939]
])

data_sizes = ['10K', '30K', '50K', '100K', '500K','1M']
x = np.arange(len(data_sizes))

mean_times = times.mean(axis=1)
std_times = times.std(axis=1, ddof=1)
plt.figure(figsize=(6, 8))
plt.errorbar(
    x, mean_times, yerr=std_times,
    fmt='o-',             
    color='black',         
    ecolor='red',          
    elinewidth=0.5,         
    markeredgewidth=0.5,     
    capsize=10          
)

plt.xticks(x, data_sizes)
plt.xlabel("Data Size (Number of Cells)")
plt.ylabel("Execution Time (h)")
plt.title("Execution Time by Data Size")

plt.grid(False)
plt.savefig('./time.pdf', dpi=300)
```
